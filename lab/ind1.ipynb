## Вариант 7
Ряды Тейлора для функции экспоненты.

```python
import math

def taylor_exp(x, n_terms=17):
    sum_exp = 0
    for n in range(n_terms):
        term = (x**n) / math.factorial(n)
        sum_exp += term
    return sum_exp

x_val = float(input("Введите значение x = "))
print(f"Приближенное значение e^{x_val}: {taylor_exp(x_val)}")
print(f"Точное значение e^{x_val}: {math.exp(x_val)}")
```
