```python
from google.colab import drive
drive.mount('/content/drive')

file_path = "/content/drive/MyDrive/Colab Notebooks/my_text_file.txt"
with open(file_path, "w") as f:
  f.write("Это текстовый файл, сохранённый в Google Drive. \n")
  f.write("Вторая строка текств.")

print("Файл успешно сохранён в Google Drive.")

with open(file_path, "r") as f:
  content = f.read()
  print("Содержимое файла: \n", content)

students = [
    ["Иванов И.И.", 20, "ИВТ-101"],
    ["Сидоров А.И.", 19, "ИВТ-105"],
    ["Михайлов С.В.", 20, "ИВТ-102"],
    ["Мастеров В.С.", 21, "ИВТ-101"],
    ["Фломастеров С.М.", 18, "ИВТ-105"]
]

csv_path = "/content/drive/MyDrive/Colab Notebooks/students.csv"
with open(csv_path, "w") as f:
  for student in students:
    f.write(",".join(map(str, student)) + "\n")

print("Файл students.csv успешно сохранён в Google Drive.")
```
